```python
import seaborn as sns

# import dataset from seaborn
titanic = sns.load_dataset('titanic')

# handle missing values or NaN
# filling in using interpolation
titanic["age"] = titanic.age.interpolate("nearest")
# filling in using specific value
titanic["embark_town"] = titanic.embark_town.fillna("Southampton")
titanic.info()
```

```python
# handle missing values by dropping colummns or rows with missing values
# axis = 1, refer to column
# axis = 0, refer to row
dropped_titanic = titanic.dropna(axis=1)
dropped_titanic
```

```python
grouped_titanic = dropped_titanic.groupby("survived")
# 1 = survived, 0 = not_survived
for group, content in grouped_titanic:
    print(group, content)
```

```python
import pandas as pd
result_list = []

for group, content in grouped_titanic:
    result_dict = {}

    if group == 1:
        result_dict["survival_status"] = "Survived"
    else:
        result_dict["survival_status"] = "Not Survived"

    for who in content["who"]:
        if who not in result_dict.keys():
            result_dict[who] = 1
        else:
            result_dict[who] += 1
    result_list.append(result_dict)
  
result_df = pd.DataFrame(result_list)
result_df
```

```python
for rdict in result_list:
    for k, v in rdict.items():
        if k == "survival_status" and v == 0:
            rdict["survival_status"] = "Not Survived"
        elif k == "survival_status" and v == 1:
            rdict["survival_status"] = "Survived"
```
