```python
import pandas as pd

covid = pd.read_csv("covid_19.csv")
covid
```

```python
covid.info()
```

```python
covid[covid["population"].isna()]
```

```python
covid.describe()
```

```python
grp_continent = covid.groupby("continent")

covid_list = []

for grp, grp_info in grp_continent:
    covid_dict = {}

    covid_dict["Continent"] = grp
    covid_dict["Country"] = grp_info.country.count()
    covid_dict["Population"] = grp_info.population.sum()
    covid_dict["Cases"] = grp_info.Cases.sum()
    covid_dict["Recovered"] = grp_info.Recovered.sum()
    covid_dict["Deaths"] = grp_info.Deaths.sum()
    covid_dict["Tests"] = grp_info.Tests.sum()
    
    covid_list.append(covid_dict)

covid_df = pd.DataFrame(covid_list)
covid_df
```

```python
covid_df = covid_df.drop(1)
covid_df
```

```python
covid_df = covid_df.reset_index(drop=True)
covid_df
```

```python
covid_df.corr(numeric_only=True)
```

```python
# There is a correlation of 0.483042 between Population and Cases. There is a positive relationship between Population and Cases, whereby the higher the number of Population, the higher the number of Cases
# There is a significant positive correlation between Cases and Tests of 0.987. In which, the higher the number of tests administered, the higher the number of cases discovered
```
