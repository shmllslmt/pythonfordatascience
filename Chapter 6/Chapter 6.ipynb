```python
data = [
    {'country': 'United Kingdom', 'median_age': 40, 'density': 200},
    {'country': 'France', 'median_age': 42, 'density': 119},
    {'country': 'Italy', 'median_age': 45, 'density': 206}
]
data_df = pd.DataFrame(data)
data_df
```

```python
new_data = pd.DataFrame([
    {'country': 'Spain', 'median_age': 40, 'density': 200},
    {'country': 'Germany', 'median_age': 42, 'density': 119},
    {'country': 'South Korea', 'median_age': 45, 'density': 206},
    {'country': 'Japan', 'median_age': 45, 'density': 206}
])
concat_df = pd.concat([data_df, new_data])
concat_df
```

```python
URL_DATA = 'https://storage.dosm.gov.my/hies/hies_state.parquet'

df = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

df
```

```python
URL_DATA = 'https://storage.dosm.gov.my/hies/hies_state_percentile.parquet'

state_df = pd.read_parquet(URL_DATA)
if 'date' in state_df.columns: state_df['date'] = pd.to_datetime(state_df['date'])

state_df
```

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic["age"] = titanic.age.interpolate("nearest")
```

```python
titanic["embark_town"] = titanic.embark_town.fillna("Southampton")
```

```python
dropped_titanic = titanic.dropna(axis=1)
dropped_titanic.info()
```
